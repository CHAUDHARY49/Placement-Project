```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('placement.csv')
```

```python
df
```

```python
df['Gender'] = [random.choice(['Male', 'Female']) for _ in range(len(df))]
```

```python
def categorize_cgpa(cgpa):
    if cgpa < 4.0:
        return 'Low'
    elif 3.0 <= cgpa < 7.0:
        return 'Medium'
    else:
        return 'High'
```

```python
df['CGPA_Category'] = df['cgpa'].apply(categorize_cgpa)
```
